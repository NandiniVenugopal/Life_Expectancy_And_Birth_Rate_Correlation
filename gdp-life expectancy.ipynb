```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array(fi['Life Expectancy'])
y = np.array(fi['GDP (current US$ 100 Million)'])
slope, intercept = np.polyfit(X, y, 1)
regression_line = slope * X + intercept
plt.scatter(X, y, label='Data Points')
plt.plot(X, regression_line, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression')
plt.grid(True)
plt.show()

A = np.vstack([X, np.ones(len(X))]).T

m, b = np.linalg.lstsq(A, y, rcond=None)[0]

print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
import statsmodels.api as sm
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Assumption 2: Independence(Durbin-Watson statistic)
import statsmodels.api as sm
from statsmodels.stats.stattools import durbin_watson
residuals = model.resid
dw_statistic = durbin_watson(residuals)
print(f'Durbin-Watson Statistic: {dw_statistic}')
```

```python
#Perform the Chi-squared goodness of fit test to check the normality assumption under type I error 5% for residuals

mean_G=np.zeros((1,1))
var_G=np.zeros((1,1))


mean_G[0,0]= np.mean(residuals)
var_G[0,0]= np.var(residuals,ddof=1)


Bin_Cutoff=np.zeros((13,1))
for i in range(13):
    for j in range(1):
        Bin_Cutoff[i,j]=sts.norm.ppf((i+1)/14,loc=mean_G[j,0],scale=np.sqrt(var_G[j,0]))


Expected_Relative_Freq=np.ones((14,1))*1/14


Observed_Absolute_Freq=np.zeros((14,1))
Observed_Absolute_Freq[0,0]=len(np.where(residuals <= Bin_Cutoff[0,0])[0])


for i in range(1,13):
    Observed_Absolute_Freq[i,0]=len(np.where(residuals <= Bin_Cutoff[i,0])[0])-np.sum(Observed_Absolute_Freq[0:i,0])
    Observed_Absolute_Freq[13,0]=len(residuals)-np.sum(Observed_Absolute_Freq[0:13,0])


Expected_Absolute_Freq=np.zeros((14,1))
Expected_Absolute_Freq[:,0]=Expected_Relative_Freq[:,0]*len(residuals)



print('p-value is', sts.chisquare(Observed_Absolute_Freq[:,0],Expected_Absolute_Freq[:,0],ddof=2)[1])
print('null hypothesis for residuals should be rejected, as p-value < 5% ')
```
